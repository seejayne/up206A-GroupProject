# Re-importing leading zeros

As expected, the FIPS, State, and County columns are missing their zeros. This took us a little bit of time to fix because it seems the intial soruce file we downloaded from Social Explorer did not include the zeroes either. We ended up formatting the columns with a leading zero in Excel, and then re-imported the file. We also specified the data type for these columns would be strings so it wouldn't clip our zeroes.

```python
df = pd.read_csv(
    'data/CensusData-Clean.csv',
    dtype=
    {
        'Geo_FIPS':str,
        'Geo_STATE':str,
        'Geo_COUNTY': str
    }
)
```

```python
df.head(5)
```

```python
df.info
```

# Import geopandas and geometry

Now it's time to get our data frames read to make some maps! First we import the geopandas library and then the geojson file from the LA Times.

```python
import geopandas as gpd
```

```python
# read in a geojson file downloaded from the LA Times
tracts=gpd.read_file('data/census-tracts-2012.geojson')
tracts.head()
```

```python
# tell me more about this dataset
tracts.info(verbose=True, null_counts=True)
```

```python
# we only really need FIPS and geometry, so let's subset the data
tracts = tracts[['name','geometry']]
tracts.head()
```

```python
# rename the columns
tracts.columns = ['FIPS','geometry']
```

```python
tracts.head()
```

```python
tracts.plot(figsize=(12,10))
```

# Merge the data

Now we want to combine the data frames together so we can add the geometry to our data

```python
# create a new dataframe based on the join
tracts_stats=tracts.merge(df,on="FIPS")
```

```python
# what does it look like now?
tracts_stats.head()
```

# Make a Map

And finally we can make a map with some data on it! Here we start with mapping the Rent Burdened Units.

```python
tracts_stats.plot(figsize=(30,30),
                 column='Rent_burdened_Units',
                 legend=True, 
                 scheme='NaturalBreaks')
```
