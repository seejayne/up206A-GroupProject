# Histogram of Rent Burdened Units in L.A County Census Tracts

```python
df['Rent_burdened_Units'].plot.hist(bins=30)
```

# Sorting Census Tracts by Highest Median Income

```python
df_sorted= df.sort_values(by='Median_income_2018',ascending = False)
```

```python
df_sorted[['Geo_NAME','Median_income_2018']].head(10)
```

```python
df_sorted.head(10).plot.bar(x='Geo_NAME',
                            y='Median_income_2018', 
                            title='Top 10 Census Tracts with Highest Median Income in Los Angeles County')
```

```python
df_povertysort= df.sort_values(by='Doing Poorly',ascending = False)
```

```python
df_povertysort.head(10).plot.bar(x='Geo_NAME',
                            y='Doing Poorly', 
                            title='10 Census Tracts with Highest Poverty level in Los Angeles County')
```
